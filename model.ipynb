## Importing all libraries

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D, BatchNormalization
```

## Loading MNIST Dataset
 

```python
mnist = tf.keras.datasets.mnist
```

## Dividing dataset into Training and Test Data

```python
(X_train,Y_train),(X_test,Y_test) = mnist.load_data()
```

## Exploring Data


```python
print(X_train.shape)        #6000 images each image represented as 28*28 matrix in training data
print(Y_train.shape)        #10000 images each image represented as 28*28 matrix in test data
print(X_test.shape)         #Array of 60000 size representing label of 60000 images
print(Y_test.shape)         #Array of 10000 size representing label of 10000 images
```

## Plotting some data from training data

```python
for i in range (0,5):
  plt.imshow(X_train[i],cmap=plt.cm.binary)
  plt.show()
  print("Label for above image: ",Y_train[i])
  print()
```

## Normalising the pixel range 
As the pixel is from 0 to 255 the pixel should be normalize and converted to values between 0 to 1 for making it suitable to input in neural network

```python
#Before Normalising the pixel values
print(X_train[0])
```

```python
#Normalizing Dataset
X_train = tf.keras.utils.normalize(X_train,axis=1)
X_test = tf.keras.utils.normalize(X_test,axis=1)
```

```python
#After Normalizing the dataset values
print(X_train[0])
```

#Resizing data to make it suitable for input in CNN layers

```python
IMG_SIZE=28
X_trainr=X_train
X_testr=X_test
X_train = np.array(X_trainr).reshape(-1,IMG_SIZE,IMG_SIZE,1)    #Converting 3d to 4d matrix of dimensions 60000*28*28*1
X_test = np.array(X_testr).reshape(-1,IMG_SIZE,IMG_SIZE,1)      #Converting 3d to 4d matrix of dimensions 10000*28*28*1
print("Training Sample Data: ",X_train.shape)
print("Test Sample Data: ",X_test.shape)
```

## Creating a CNN Deep Neural Networks layer and joining them

```python
model = Sequential()

#First Convolutional Layer
model.add(Conv2D(64,(3,3),input_shape = X_train.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

#Second Convolution Layer
model.add(Conv2D(64,(3,3)))                 #2nd convolution layer
model.add(Activation("relu"))               #activation Function
model.add(MaxPooling2D(pool_size=(2,2)))    #Maxpooling Function

#Third Convolution Layer
model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))


#Fully Connected Layer1
model.add(Flatten())
model.add(Dense(64))
model.add(Activation("relu"))

#Fully Connected Layer2
model.add(Dense(32))
model.add(Activation("relu"))

#Fully Connected Layer3
model.add(Dense(10))  #Give output in classes from 0 to 10
model.add(Activation("softmax"))  #Softmax function gives probability for each class


```

```python
model.summary()  #Summary of above model
```

```python
#Compiling the above model
model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=['accuracy'])
```

```python
#Training the model
model.fit(X_train, Y_train,epochs=10, validation_data=(X_test,Y_test))
```

```python
#Evaluating model
test_loss, test_acc = model.evaluate(X_test,Y_test)
print("Test Loss: ",test_loss*100)
print("Test Accuracy: ",test_acc*100)
```

```python
#storing predictions of test data
predictions = model.predict(X_test)
```

```python
def plot_images_sample(Y, X):
    # Draw plot for images sample
    
    plt.figure(figsize=(10,10))
    rand_indicies = np.random.randint(len(X), size=20)
    for i in range(20):
        plt.subplot(5,5,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        index = rand_indicies[i]
        plt.imshow(np.squeeze(X[index]), cmap=plt.cm.binary)
        plt.xlabel(Y[index])
    plt.show()
```

```python
plot_images_sample(np.argmax(predictions,axis=1),X_testr)
```

```python
cnt=0
for i in range(0,len(X_testr)):
  X_data=X_testr[i]
  x=np.argmax(predictions[i])
  if(x!=Y_test[i]):
    cnt+=1
    plt.imshow(X_data)
    plt.show()
    print("Prediction value: ",x,"Original Value: ",Y_test[i])

print("Total Miss Values: ",cnt)
```
